```python
def display_items(items):
    """자판기 상품 목록과 가격을 출력합니다."""
    print("\n--- 자판기 상품 목록 ---")
    for item, price in items.items():
        print(f"{item}: {price}원")
    print("----------------------")

def insert_money(current_balance):
    """사용자가 돈을 넣는 기능을 처리합니다."""
    while True:
        try:
            money = int(input(f"현재 잔액: {current_balance}원. 돈을 넣어주세요 (원 단위, 종료하려면 0): "))
            if money < 0:
                print("음수는 입력할 수 없습니다.")
            elif money == 0:
                print("돈 넣기를 취소합니다.")
                return current_balance # 돈 넣기 취소 시 현재 잔액 유지
            else:
                current_balance += money
                print(f"{money}원이 입금되었습니다. 현재 잔액: {current_balance}원")
                return current_balance
        except ValueError:
            print("유효한 금액(숫자)을 입력해주세요.")

def select_item(current_balance, items):
    """사용자가 상품을 선택하고 구매를 처리합니다."""
    while True:
        item_choice = input("구매할 상품 이름을 입력하세요 (취소: '취소'): ").strip()

        if item_choice == '취소':
            print("상품 선택을 취소합니다.")
            return current_balance

        if item_choice in items:
            item_price = items[item_choice]
            if current_balance >= item_price:
                current_balance -= item_price
                print(f"'{item_choice}'가(이) 나왔습니다. 감사합니다!")
                return current_balance
            else:
                print(f"잔액이 부족합니다. '{item_choice}'는 {item_price}원 입니다. 현재 잔액: {current_balance}원")
                # 잔액이 부족하면 다시 선택하거나 돈을 더 넣도록 유도
                action = input("돈을 더 넣으시겠습니까? (예/아니오): ").lower()
                if action == '예':
                    return current_balance # 돈을 더 넣는 선택지로 돌아가도록 현재 잔액을 유지하고 반환
                else:
                    return current_balance # 다시 상품 선택으로 돌아가도록 현재 잔액 유지
        else:
            print("선택하신 상품은 없습니다. 다시 확인해주세요.")
            display_items(items) # 없는 상품을 선택했을 때 다시 목록 보여주기

def return_change(current_balance):
    """잔돈을 반환하고 잔액을 0으로 초기화합니다."""
    if current_balance > 0:
        print(f"잔돈 {current_balance}원이 반환되었습니다. 감사합니다!")
    else:
        print("반환할 잔돈이 없습니다.")
    return 0 # 잔액을 0으로 초기화

def main():
    """자판기 프로그램의 메인 함수입니다."""
    items = {
        "콜라": 1000,
        "사이다": 1000,
        "환타": 900,
        "생수": 500,
        "초코우유": 700
    }
    current_balance = 0

    print("--- 파이썬 자판기에 오신 것을 환영합니다! ---")

    while True:
        display_items(items)
        print(f"현재 잔액: {current_balance}원")
        print("\n--- 메뉴 ---")
        print("1. 돈 넣기")
        print("2. 상품 선택")
        print("3. 잔돈 반환")
        print("4. 종료")
        print("------------")

        choice = input("원하는 동작을 선택하세요 (1-4): ").strip()

        if choice == '1':
            current_balance = insert_money(current_balance)
        elif choice == '2':
            current_balance = select_item(current_balance, items)
        elif choice == '3':
            current_balance = return_change(current_balance)
        elif choice == '4':
            print("자판기를 종료합니다. 안녕히 가세요!")
            if current_balance > 0:
                current_balance = return_change(current_balance) # 종료 전 잔돈 반환
            break
        else:
            print("잘못된 선택입니다. 1에서 4 사이의 숫자를 입력해주세요.")

# 프로그램 실행
if __name__ == "__main__":
    main()
```
